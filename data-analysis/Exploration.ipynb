```r
d <- read.csv("comb-data.csv")
d$coverage_inc <- d$edges_covered - d$initial_coverage
d$cif <- factor(d$coverage_inc, ordered=TRUE)

```

```r
ranks = c()
for (rown in 1:nrow(d)) {
    rank = 1
    for (rown_inner in 1:nrow(d)) {
        if (d[rown, "initial_coverage"] == d[rown_inner, "initial_coverage"] &&
            d[rown, "number.of.seeds"] == d[rown_inner, "number.of.seeds"] &&
            d[rown, "fuzzer"] != d[rown_inner, "fuzzer"]) {
            if (d[rown, "coverage_inc"] < d[rown_inner, "coverage_inc"]) {
                rank <- rank + 1
            }
        }
    }
    ranks[rown] = rank
}
d$ranks <- factor(ranks, ordered=TRUE)
```

```r
# Assumption 2 (ANCOVA): No interaction between covariate and grouping variable
d %>% anova_test(coverage_inc ~ fuzzer*number.of.seeds)
d %>% anova_test(coverage_inc ~ fuzzer*mean.seed.size)
d %>% anova_test(coverage_inc ~ fuzzer*initial_coverage)
```

```r
# Fit the model, the covariate goes first
model <- lm(coverage_inc ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
```

```r
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)
```

```r
# Assumption 3 (ANCOVA): output is normally distributed

# Fit the model, the covariate goes first
model <- lm(coverage_inc ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)

# Fit the model, the covariate goes first
modelb <- lm(coverage_inc ~ mean.seed.size + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelb.metrics <- augment(modelb) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelb.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelb.metrics$.resid)


# Fit the model, the covariate goes first
modelc <- lm(coverage_inc ~ initial_coverage + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelc.metrics <- augment(modelc) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelc.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelc.metrics$.resid)
```

```r
model <- polr(formula = ranks ~ number.of.seeds + mean.seed.size + initial_coverage + initial_coverage*mean.seed.size, data = d)
summary(model)
```

```r

# get coefficients (it's in matrix form)
coefficients <- summary(model)$coefficients

# calculate p-values
p_value <- (1 - pnorm(abs(coefficients[ ,"t value"]), 0, 1))*2

# bind back to coefficients
(coefficients <- cbind(coefficients, p_value))

# calculate odds ratios
odds_ratio <- exp(coefficients[ ,"Value"])

# combine with coefficient and p_value
(coefficients <- cbind(
  coefficients[ ,c("Value", "p_value")],
  odds_ratio
))
```

```r
generalhoslem::lipsitz.test(model)
```
