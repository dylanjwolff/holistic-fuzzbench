```r
# install.packages("tidyverse")
# install.packages("ggpubr")
# install.packages("rstatix")
# install.packages("broom")
library(tidyverse)
library(ggpubr)
library(rstatix)
library(broom)
```

```r
# Assumption 1 (ANCOVA): Visually check for  linear relationship between covariate and output 
ggscatter(
  d, x = "number.of.seeds", y = "edges_covered",
  color = "fuzzer", add = "reg.line"
  )+
  stat_regline_equation(
    aes(label =  paste(..eq.label.., ..rr.label.., sep = "~~~~"), color = fuzzer)
    )
ggscatter(
  d, x = "mean.seed.size", y = "edges_covered",
  color = "fuzzer", add = "reg.line"
  )+
  stat_regline_equation(
    aes(label =  paste(..eq.label.., ..rr.label.., sep = "~~~~"), color = fuzzer)
    )
ggscatter(
  d, x = "initial_coverage", y = "edges_covered",
  color = "fuzzer", add = "reg.line"
  )+
  stat_regline_equation(
    aes(label =  paste(..eq.label.., ..rr.label.., sep = "~~~~"), color = fuzzer)
    )
```

```r
# Assumption 2 (ANCOVA): No interaction between covariate and grouping variable
d %>% anova_test(edges_covered ~ fuzzer*number.of.seeds)
d %>% anova_test(edges_covered ~ fuzzer*mean.seed.size)
d %>% anova_test(edges_covered ~ fuzzer*initial_coverage)
```

```r
# Fit the model, the covariate goes first
model <- lm(edges_covered ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
```

```r
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)
```

```r
# Assumption 3 (ANCOVA): output is normally distributed

# Fit the model, the covariate goes first
model <- lm(edges_covered ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)

# Fit the model, the covariate goes first
modelb <- lm(edges_covered ~ mean.seed.size + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelb.metrics <- augment(modelb) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelb.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelb.metrics$.resid)


# Fit the model, the covariate goes first
modelc <- lm(edges_covered ~ initial_coverage + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelc.metrics <- augment(modelc) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelc.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelc.metrics$.resid)
```
