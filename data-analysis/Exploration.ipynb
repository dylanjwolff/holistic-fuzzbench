```r
d <- read.csv("comb-data.csv")
```

```r
plot(d$number.of.seeds, d$edges_covered, col=as.factor(d$fuzzer))
lmout <- lm(edges_covered ~ number.of.seeds,data=d)
abline(lmout)
```

```r
plot(d$mean.seed.size, d$edges_covered, col=as.factor(d$fuzzer))
lmout <- lm(edges_covered ~ mean.seed.size,data=d)
abline(lmout)
```

```r
lm(edges_covered ~ mean.seed.size + number.of.seeds,data=d)
```

```r
install.packages("tidyverse")
install.packages("ggpubr")
install.packages("rstatix")
install.packages("broom")
library(tidyverse)
library(ggpubr)
library(rstatix)
library(broom)
```

```r
ggscatter(
  d, x = "number.of.seeds", y = "edges_covered",
  color = "fuzzer", add = "reg.line"
  )+
  stat_regline_equation(
    aes(label =  paste(..eq.label.., ..rr.label.., sep = "~~~~"), color = fuzzer)
    )
```

```r
d %>% anova_test(edges_covered ~ fuzzer*number.of.seeds)
```

```r
# Fit the model, the covariate goes first
model <- lm(edges_covered ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
```

```r
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)
```

```r
model.metrics %>% levene_test(.resid ~ fuzzer)
```

```r
res.aov <- d %>% anova_test(edges_covered ~ number.of.seeds + fuzzer)
get_anova_table(res.aov)
```
