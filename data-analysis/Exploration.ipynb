```r
d$coverage_increase <- d$edges_covered - d$initial_coverage
afld <- d[d$fuzzer == "afl",]
lfd <- d[d$fuzzer == "libfuzzer",]
lfd$trial_id <- lfd$trial_id - 40
coverage_increase <- afld$coverage_increase - lfd$coverage_increase
differing_res <- data.frame(coverage_increase)
differing_res$trial_id <- afld$trial_id
differing_res$initial_coverage <- afld$initial_coverage
differing_res$mean.seed.size <- afld$mean.seed.size
differing_res$number.of.seeds <- afld$number.of.seeds
```

```r
# Assumption 2 (ANCOVA): No interaction between covariate and grouping variable
d %>% anova_test(coverage_increase ~ fuzzer*number.of.seeds)
d %>% anova_test(coverage_increase ~ fuzzer*mean.seed.size)
d %>% anova_test(coverage_increase ~ fuzzer*initial_coverage)
```

```r
# Assumption 3 (ANCOVA): output is normally distributed

# Fit the model, the covariate goes first
model <- lm(coverage_increase ~ number.of.seeds + fuzzer, data = d)
# Inspect the model diagnostic metrics
model.metrics <- augment(model) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(model.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(model.metrics$.resid)

# Fit the model, the covariate goes first
modelb <- lm(coverage_increase ~ mean.seed.size + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelb.metrics <- augment(modelb) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelb.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelb.metrics$.resid)


# Fit the model, the covariate goes first
modelc <- lm(coverage_increase ~ initial_coverage + fuzzer, data = d)
# Inspect the model diagnostic metrics
modelc.metrics <- augment(modelc) %>%
  select(-.hat, -.sigma, -.fitted) # Remove details
head(modelc.metrics, 3)
# Not normal! so, as expected seems we need non-parametric
shapiro_test(modelc.metrics$.resid)
```

```r
# Assumption 4 (ANCOVA): variances are approximately equal
model.metrics %>% levene_test(.resid ~ fuzzer)
modelb.metrics %>% levene_test(.resid ~ fuzzer)
modelc.metrics %>% levene_test(.resid ~ fuzzer)
```
